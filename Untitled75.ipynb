```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def removeNthFromEnd(head, n):
    dummy = ListNode(0)
    dummy.next = head
    first = dummy
    second = dummy

    # Move the first pointer n steps ahead
    for _ in range(n + 1):
        first = first.next

    # Move first to the end, maintaining the gap
    while first is not None:
        first = first.next
        second = second.next

    # Remove the nth node from the end
    second.next = second.next.next

    return dummy.next

# Helper function to print the linked list
def printLinkedList(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

# Example usage:
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)
n = 2

print("Input:")
print("Linked List:")
printLinkedList(head)
print("n:", n)

new_head = removeNthFromEnd(head, n)

print("\nOutput:")
printLinkedList(new_head)
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def getIntersectionNode(headA, headB):
    if not headA or not headB:
        return None

    # Function to get the length of a linked list
    def getLength(node):
        length = 0
        while node:
            length += 1
            node = node.next
        return length

    # Get the length of both linked lists
    lengthA = getLength(headA)
    lengthB = getLength(headB)

    # Adjust the longer list's pointer to match the length of the shorter list
    while lengthA > lengthB:
        headA = headA.next
        lengthA -= 1
    while lengthB > lengthA:
        headB = headB.next
        lengthB -= 1

    # Traverse both lists until intersection point is found
    while headA != headB:
        headA = headA.next
        headB = headB.next

    return headA

# Example usage:
# Construct the example lists
list1 = ListNode(1)
list1.next = ListNode(2)
list1.next.next = ListNode(3)
list1.next.next.next = ListNode(4)

list2 = ListNode(9)
list2.next = ListNode(8)
list2.next.next = list1.next.next  # Intersection point

print("Input:")
print("List 1:")
printLinkedList(list1)
print("List 2:")
printLinkedList(list2)

intersection_node = getIntersectionNode(list1, list2)

print("\nOutput:")
if intersection_node:
    print("Node with value", intersection_node.val)
else:
    print("No intersection found")
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def addTwoNumbers(l1, l2):
    dummy = ListNode()
    current = dummy
    carry = 0

    while l1 or l2 or carry:
        # Calculate the sum of the current digits and the carry
        sum_val = carry
        if l1:
            sum_val += l1.val
            l1 = l1.next
        if l2:
            sum_val += l2.val
            l2 = l2.next

        # Update carry and create a new node with the sum value
        carry = sum_val // 10
        sum_val %= 10
        current.next = ListNode(sum_val)
        current = current.next

    return dummy.next

# Helper function to print the linked list
def printLinkedList(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

# Example usage:
list1 = ListNode(2)
list1.next = ListNode(4)
list1.next.next = ListNode(3)

list2 = ListNode(5)
list2.next = ListNode(6)
list2.next.next = ListNode(4)

print("Input:")
print("List 1:")
printLinkedList(list1)
print("List 2:")
printLinkedList(list2)

result = addTwoNumbers(list1, list2)

print("\nOutput:")
printLinkedList(result)
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def rotateRight(head, k):
    if not head or not head.next or k == 0:
        return head

    # Calculate the length of the linked list
    length = 1
    current = head
    while current.next:
        length += 1
        current = current.next

    # Calculate the effective rotation value
    k %= length

    # If k is 0 after taking the modulo, no rotation is needed
    if k == 0:
        return head

    # Find the (length - k)th node
    rotate_index = length - k
    current.next = head  # Connect the end of the list to the head to form a loop
    for _ in range(rotate_index):
        current = current.next

    # Set the new head and break the loop
    new_head = current.next
    current.next = None

    return new_head

# Helper function to print the linked list
def printLinkedList(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

# Example usage:
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)
k = 2

print("Input:")
print("Linked List:")
printLinkedList(head)
print("k:", k)

head = rotateRight(head, k)

print("\nOutput:")
printLinkedList(head)

```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def rearrangeLinkedList(head):
    if not head or not head.next:
        return head

    odd_head = ListNode()
    even_head = ListNode()

    odd_current = odd_head
    even_current = even_head

    current = head
    count = 1

    while current:
        if count % 2 == 1:  # Odd position
            odd_current.next = current
            odd_current = odd_current.next
        else:  # Even position
            even_current.next = current
            even_current = even_current.next

        current = current.next
        count += 1

    odd_current.next = even_head.next  # Append the even positioned nodes list to the end of the odd positioned nodes list
    even_current.next = None  # Set the end of even positioned nodes list to None to avoid cycles

    return odd_head.next

# Helper function to print the linked list
def printLinkedList(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

# Example usage:
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

print("Input:")
printLinkedList(head)

head = rearrangeLinkedList(head)

print("\nOutput:")
printLinkedList(head)
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def addOne(head):
    # Helper function to reverse a linked list
    def reverseLinkedList(head):
        prev = None
        current = head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        return prev

    # Reverse the linked list
    reversed_head = reverseLinkedList(head)

    current = reversed_head
    carry = 1

    while current:
        total = current.val + carry
        current.val = total % 10
        carry = total // 10
        if carry == 0:
            break
        if not current.next:
            current.next = ListNode(carry)
            break
        current = current.next

    # Reverse the linked list back to its original order
    return reverseLinkedList(reversed_head)

# Helper function to print the linked list
def printLinkedList(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

# Example usage:
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)

print("Input:")
printLinkedList(head)

head = addOne(head)

print("\nOutput:")
printLinkedList(head)
```

```python
def searchInsert(nums, target):
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = (left + right) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    # If the target is not found, left will be the index where it should be inserted
    return left

# Example usage:
nums = [1, 3, 5, 6]
target = 5
print("Input:")
print("nums =", nums)
print("target =", target)

result = searchInsert(nums, target)

print("\nOutput:")
print("Index where the target is found or should be inserted:", result)
```

```python
def findMin(nums):
    left = 0
    right = len(nums) - 1

    while left < right:
        mid = (left + right) // 2

        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid

    # At the end of the loop, 'left' will be pointing to the minimum element
    return nums[left]

# Example usage:
nums = [4, 5, 6, 7, 0, 1, 2]
print("Input:")
print("nums =", nums)

result = findMin(nums)

print("\nOutput:")
print("Minimum element in the rotated sorted array:", result)
```

```python
def search(nums, target):
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = (left + right) // 2

        if nums[mid] == target:
            return mid

        # Check if the left half is sorted
        if nums[left] <= nums[mid]:
            # If the target is within the sorted left half
            if nums[left] <= target < nums[mid]:
                right = mid - 1
            else:
                left = mid + 1
        # Otherwise, the right half is sorted
        else:
            # If the target is within the sorted right half
            if nums[mid] < target <= nums[right]:
                left = mid + 1
            else:
                right = mid - 1

    return -1  # Target not found

# Example usage:
nums = [4, 5, 6, 7, 0, 1, 2]
target = 0
print("Input:")
print("nums =", nums)
print("target =", target)

result = search(nums, target)

print("\nOutput:")
print("Index of the target value in the rotated sorted array:", result)
```

```python
def findPeakElement(nums):
    left = 0
    right = len(nums) - 1

    while left < right:
        mid = (left + right) // 2

        # If the element to the right of mid is greater than mid, there must be a peak to the right
        if nums[mid] < nums[mid + 1]:
            left = mid + 1
        # Otherwise, if the element to the left of mid is greater than mid, there must be a peak to the left
        elif nums[mid] > nums[mid + 1]:
            right = mid
        # If neither condition is true, mid is at a peak
        else:
            return mid

    # At the end of the loop, 'left' will be pointing to the peak element
    return left

# Example usage:
nums = [1, 2, 3, 1]
print("Input:")
print("nums =", nums)

result = findPeakElement(nums)

print("\nOutput:")
print("Index of the peak element:", result)
```

```python
def countNegatives(grid):
    m = len(grid)  # Number of rows
    n = len(grid[0])  # Number of columns

    count = 0
    row = 0
    col = n - 1  # Start from the rightmost column of the first row

    while row < m and col >= 0:
        if grid[row][col] < 0:
            # If the current element is negative, then all the elements to its left in the same row will also be negative
            count += (m - row)
            col -= 1  # Move left within the same row
        else:
            row += 1  # Move down to the next row
        
    return count

# Example usage:
grid = [
    [4, 3, 2, -1],
    [3, 2, 1, -1],
    [1, 1, -1, -2],
    [-1, -1, -2, -3]
]
print("Input:")
for row in grid:
    print(row)

result = countNegatives(grid)

print("\nOutput:")
print("Number of negative numbers in the matrix:", result)
```

```python
def searchMatrix(matrix, target):
    if not matrix or not matrix[0]:
        return False
    
    rows = len(matrix)
    cols = len(matrix[0])
    
    left, right = 0, rows * cols - 1
    
    while left <= right:
        mid = left + (right - left) // 2
        num = matrix[mid // cols][mid % cols]
        
        if num == target:
            return True
        elif num < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return False

# Example usage:
matrix = [
    [1, 3, 5, 7],
    [10, 11, 16, 20],
    [23, 30, 34, 60]
]
target = 3
print("Input:")
print("matrix =", matrix)
print("target =", target)

result = searchMatrix(matrix, target)

print("\nOutput:")
print("Is the target value present in the matrix?", result)
```

```python
def findMedianSortedArrays(nums1, nums2):
    merged = []
    i = j = 0

    while i < len(nums1) and j < len(nums2):
        if nums1[i] < nums2[j]:
            merged.append(nums1[i])
            i += 1
        else:
            merged.append(nums2[j])
            j += 1

    while i < len(nums1):
        merged.append(nums1[i])
        i += 1

    while j < len(nums2):
        merged.append(nums2[j])
        j += 1

    total_length = len(nums1) + len(nums2)
    if total_length % 2 == 0:
        return (merged[total_length // 2 - 1] + merged[total_length // 2]) / 2
    else:
        return merged[total_length // 2]

# Example usage:
nums1 = [1, 3]
nums2 = [2]
print(findMedianSortedArrays(nums1, nums2))  # Output: 2.0
```

```python
def nextGreatestLetter(letters, target):
    left = 0
    right = len(letters) - 1

    while left <= right:
        mid = left + (right - left) // 2

        if letters[mid] <= target:
            left = mid + 1
        else:
            right = mid - 1

    return letters[left % len(letters)]

# Example usage:
letters = ['c', 'f', 'j']
target = 'a'
print(nextGreatestLetter(letters, target))  # Output: 'c'
```

```python
def sortColors(nums):
    left = 0
    right = len(nums) - 1
    i = 0

    while i <= right:
        if nums[i] == 0:
            nums[i], nums[left] = nums[left], nums[i]
            left += 1
            i += 1
        elif nums[i] == 2:
            nums[i], nums[right] = nums[right], nums[i]
            right -= 1
        else:
            i += 1

# Example usage:
nums = [2, 0, 2, 1, 1, 0]
sortColors(nums)
print(nums)  # Output: [0, 0, 1, 1, 2, 2]
```

```python
import random

def findKthLargest(nums, k):
    def partition(left, right, pivot_index):
        pivot = nums[pivot_index]
        nums[pivot_index], nums[right] = nums[right], nums[pivot_index]
        store_index = left

        for i in range(left, right):
            if nums[i] < pivot:
                nums[i], nums[store_index] = nums[store_index], nums[i]
                store_index += 1

        nums[right], nums[store_index] = nums[store_index], nums[right]
        return store_index

    def select(left, right, k_smallest):
        if left == right:
            return nums[left]

        pivot_index = random.randint(left, right)
        pivot_index = partition(left, right, pivot_index)

        if k_smallest == pivot_index:
            return nums[k_smallest]
        elif k_smallest < pivot_index:
            return select(left, pivot_index - 1, k_smallest)
        else:
            return select(pivot_index + 1, right, k_smallest)

    return select(0, len(nums) - 1, len(nums) - k)

# Example usage:
nums = [3, 2, 1, 5, 6, 4]
k = 2
print(findKthLargest(nums, k))  # Output: 5
```

```python
def wiggleSort(nums):
    nums.sort()
    for i in range(1, len(nums) - 1, 2):
        nums[i], nums[i + 1] = nums[i + 1], nums[i]

# Example usage:
nums = [3, 5, 2, 1, 6, 4]
wiggleSort(nums)
print(nums)  # Output: [3, 5, 1, 6, 2, 4]
```

```python
def factorial(n):
    if n == 0:
        return 1
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Example usage:
num = 5
print(factorial(num))  # Output: 120
```

```python
import math

def isPrime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# Example usage:
num = 7
print(isPrime(num))  # Output: True
```

```python
def fibonacci(n):
    fib_series = [0, 1]
    while fib_series[-1] + fib_series[-2] <= n:
        fib_series.append(fib_series[-1] + fib_series[-2])
    return fib_series

# Example usage:
n = 8
print(fibonacci(n))  # Output: [0, 1, 1, 2, 3, 5, 8]
```

```python
def power(base, exponent):
    if exponent == 0:
        return 1
    elif exponent == 1:
        return base
    else:
        return base * power(base, exponent - 1)

# Example usage:
base = 3
exponent = 4
print(power(base, exponent))  # Output: 81
```
